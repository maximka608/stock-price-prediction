```python
import yfinance as yf
import pandas as pd
import ta
```

# AAPL

```python
symbol = "AAPL"
historical_data_day_tf = yf.download(symbol, start="2010-01-01", end='2025-02-06')
```

```python
historical_data_day_tf.head()
```

```python
historical_data_day_tf.info()
```

```python
historical_data_day_tf.columns = historical_data_day_tf.columns.get_level_values(0)
```

```python
historical_data_day_tf = historical_data_day_tf[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
historical_data_day_tf
```

```python
def get_indicators(df_daily):
    df_daily["SMA"] = ta.trend.sma_indicator(df_daily["Close"], window=50)
    df_daily["EMA"] = ta.trend.ema_indicator(df_daily["Close"], window=50)
    df_daily["RSI"] = ta.momentum.rsi(df_daily["Close"], window=14)
    df_daily["MACD"] = ta.trend.macd(df_daily["Close"])
    df_daily["MACD_Signal"] = ta.trend.macd_signal(df_daily["Close"])
    df_daily["BB_Upper"] = ta.volatility.bollinger_hband(df_daily["Close"], window=20, window_dev=2)
    df_daily["BB_Lower"] = ta.volatility.bollinger_lband(df_daily["Close"], window=20, window_dev=2)
```

```python
get_indicators(historical_data_day_tf)
```

```python
historical_data_day_tf.to_csv('AAPL_day.csv')
```
